```python
#load the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

```

load the dataset

```python

train= pd.read_csv('/content/sign_mnist.csv')
labels=train["label"]
Y=train.iloc[0:,1:]
X_train, X_test, Y_train, Y_test = train_test_split(Y, labels, test_size=0.20, random_state=33)
```

Classifier

```python
print('Classification Report for Support Vector Machine Classifier')
svm_model= svm.SVC(kernel='rbf')
clf = svm_model.fit(X_train, Y_train)
pred_tr = svm_model.predict(X_train)
pred_te = svm_model.predict(X_test)
print(' Evaluation  Test Data ')
score = svm_model.score(X_test, Y_test)
print('Accuracy Score: ', score)
print(classification_report(Y_test, pred_te))
print('--------')
print(' Evaluation  Training Data ')
score = svm_model.score(X_train, Y_train)
print('Accuracy Score: ', score)
print(classification_report(Y_train, pred_tr))
print('------')
```

Decision Tree Classifier

```python
print('Classification Report for Decision Tree Classifier')
DecisionTreeClassifier = DecisionTreeClassifier()
dt = DecisionTreeClassifier.fit(X_train, Y_train)
pred_dttr = dt.predict(X_train)
pred_dtte = dt.predict(X_test)
print(' Evaluation  Test Data')
score = dt.score(X_test, Y_test)
print('Accuracy Score: ', score)
print(classification_report(Y_test, pred_dtte))
print('-------')
print(' Evaluation on Training Data ')
score = dt.score(X_train, Y_train)
print('Accuracy Score: ', score)
print(classification_report(Y_train, pred_dttr))
print('---------')
```

KNeighborsClassifier

```python
print('Classification Report for KNeighborsClassifier')
KNeighborsClassifier= KNeighborsClassifier(n_neighbors=5)
KNeighborsClassifier.fit(X_train, Y_train)
pred_knntr = KNeighborsClassifier.predict(X_train)
pred_knnte = KNeighborsClassifier.predict(X_test)
print('----- Evaluation on Test Data -----')
score_knnte = KNeighborsClassifier.score(X_test, Y_test)
print('Accuracy Score: ', score_knnte)
print(classification_report(Y_test, pred_knnte))
print('-')
print(' Evaluation on Training Data ')
score_knntr = KNeighborsClassifier.score(X_train, Y_train)
print('Accuracy Score: ', score_knntr)
print(classification_report(Y_train, pred_knntr))
print('--')
```
